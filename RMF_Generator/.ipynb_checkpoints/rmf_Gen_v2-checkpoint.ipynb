This program is developed to model, process, and calibrate the Response Matrix File (RMF) for the AstroSat Soft X-ray Telescope (SXT). 

The program uses analytical model capable of reproducing the RMF shape using key detector and physical parameters. It generates synthetic RMFs across a defined parameter space to study response variations and assess model behavior to determine the optimal set of parameters that can be further used for updating the RMFs.

```python
# -----------------------------------
# Plotting function
# -----------------------------------
def plot_rmf(E_channel, rmf_dist, mormf_dist, Ein, pdf,Bt, f0):
    gs = gridspec.GridSpec(2, 1, height_ratios=[2, 0.5], hspace=0)
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1], sharex=ax0)

    # --- Main RMF plot (log-log scale) ---
    ax0.plot(E_channel, rmf_dist, 'o-', color='k', ms=1.5, lw=1, alpha=0.8, label=f'SXT rmf @ {Ein:.4f} keV')
    ax0.plot(E_channel, mormf_dist, color='tomato', lw=2, label=f'model rmf Gen1; {Bt:.4f}, {f0:.4f}')
    ax0.set_xscale('log')
    ax0.set_yscale('log')
    ax0.set_xlim(0.05, 2.2)
    ax0.set_ylim(1e-6, 1e-1)
    ax0.set_ylabel('N(x) distribution')
    ax0.legend()
    ax0.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    ax0.set_title(f'SXT RMF model for X-ray Photon Energy {Ein:.4f} keV')

    # --- Residual plot (Data / Model, log-log scale) ---
    ax1.plot(E_channel, rmf_dist / mormf_dist, '-.', color='tomato', lw=0.8, label='Ratio: Data / Model')
    ax1.axhline(1, color='black', linestyle='--', linewidth=0.6)
    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.set_xlim(0.05, 2.2)
    ax1.set_ylim(0, 15)
    ax1.set_xlabel('Channel Energy (E) keV')
    ax1.set_ylabel('Data/Model')
    ax1.legend(prop={'size': 5})
    ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

    plt.tight_layout()
    pdf.savefig()
    plt.close()
```

```python
# -----------------------------------
# Main loop
# -----------------------------------
pdf_filename = os.path.join(rootdir5, 'RMF_Gen_v2_check_f0.pdf')

with PdfPages(pdf_filename) as pdf:
    for i in range(1):
        idx = 83 + i

        # Load original rmf
        data2 = np.loadtxt(f'{rootdir6}/rmf_{idx}.dat')
        E_channel = data2[:, 0] * 1e-2  # convert to keV
        rmf_dist = data2[:, 1]

        # Load attenuation coefficient
        data3 = np.loadtxt(f'{rootdir2}/attcoeff.dat')
        Ein = data3[:, 0][idx]
        mu = data3[:, 1][idx]

        print('Energy of Input X-ray Photon:', Ein)
        #Define the Parameter spacefor RMF generation
        # Load model parameters if available
        l = 0.1
        
        param_sig_file = f'{rootdir3}/sig_fano.dat'

        if not os.path.exists(param_sig_file):
            # Define parameter range manually
            sig_range = np.arange(0.01, 0.11, 0.01)   # sigma range: 0.01 â†’ 0.1 in steps of 0.01
            print("Parameter file not found. Using default sigma range:", sig_range)
        else:
            # Load model parameters from file
            data4 = np.loadtxt(param_sig_file)
            sig = data4[idx-83, 1]              # load specific sigma value
            print(f"Loaded sigma from file: {sig:.5f}")

        param_tau_file = f'{rootdir3}/l_attlen.dat'

        if not os.path.exists(param_tau_file):
            # Define parameter range manually for tau
            N = 50  # total number of points
            x = np.linspace(0, 1, N)
            tau_range = 0.01 + (10 - 0.01) * x**3  # tau varies non-linearly between 0.02 and 10
            print("Parameter file not found. Using default tau range:", tau_range)
        else:
            # Load tau parameter values from file
            data5 = np.loadtxt(param_tau_file)
            tau = data5[idx-82, 1]             # load specific tau value
            print(f"Loaded tau from file: {tau:.5f}")

        param_beta_file = f'{rootdir3}/beta_sim1_orig.dat'

        if not os.path.exists(param_beta_file):
            # Define parameter range manually for tau
            N = 50  # total number of points
            x = np.linspace(0, 1, N)
            beta_range = 0.01 + (10 - 0.01) * x**3  # tau varies non-linearly between 0.02 and 10
            print("Parameter file not found. Using default beta range:", beta_range)
        else:
            # Load tau parameter values from file
            data5 = np.loadtxt(param_beta_file)
            Bt = data5[idx-83, 1]             # load specific tau value
            print(f"Loaded beta from file: {Bt:.5f}")

        # Generate 50 f0 values 
        N= 50 
        for k, n in enumerate(np.linspace(0, 1, N), start=1):

            f0 = 0.1 + (1 - 0.01) * n**1.5

            # Print index only
#            print(f"f0 value{k}={f0}")

            # Compute model RMF distribution
            print(f"Params Test {k}: Energy={Ein}, mu={mu}, l={l}, Bt={Bt}, sig={sig}, tau={tau}, norm={norm}, f0={f0}")

            mormfdist1 = mormf_set1(E_channel, l, Bt, sig, tau, norm, f0, mu, Ein)

            # Plot and save
            plot_rmf(E_channel, rmf_dist, mormfdist1, Ein, pdf)

print(f"Plots saved to {pdf_filename}")
```
