This program is developed to model, process, and calibrate the Response Matrix File (RMF) for the AstroSat Soft X-ray Telescope (SXT). 

The program uses analytical model capable of reproducing the RMF shape using key detector and physical parameters. It generates synthetic RMFs across a defined parameter space to study response variations and assess model behavior to determine the optimal set of parameters that can be further used for updating the RMFs.

###  I. RMF Model Description and Parameterization

**Model Parameters:**  
- `Ein` â€” Input X-ray Photon Energy
- Î¼    â€” attenuation coefficient (1/Î¼m)  
- Ïƒ    â€” Energy Resolution  
- Ï„    â€” Damping coefficient (Î¼m)
- l    â€” Transition depth of energy loss functions (Î¼m)  
- fâ‚€   â€” fractional Energy loss at Surface 
- Î²    â€” Shape parameter for shoulder due to charge loss
- norm â€” normalization factor

### **III. RMF Model**

The RMF is constructed by dividing the incident energy $(E_{\rm in})$ range into three regimes.  
Each Energy regime is modeled by : `mormf_set1`, `mormf_set2`, and `mormf_set3`.

#### **`mormf_set1` â€” Set 1 (0.4175â€“1.8375 keV)**
This model describes the low-energy response where:
- the main peak deviates from an ideal Gaussian shape,
- shoulder-like features emerge.
- **primary photopeak with shoulder**

#### **`mormf_set2` â€” Set 2 (1.8425â€“5.7875 keV)**
This model represents the intermediate-energy region characterized by:
- a well-defined **primary photopeak with shoulder**,
- an **escape peak** appearing at $(E_{\rm in} - E_{\rm Si})$ keV,


#### **`mormf_set3` â€” Set 3 (5.7925â€“10.2475 keV)**
This model covers the high-energy regime exhibiting:
- the main **photopeak with shoulder**,
- an **escape peak** at $(E_{\rm in} - E_{\rm Si})$ keV,
- a **fluorescence peak** at $(E_{\rm Si})$ keV,
- additional small off-peak features associated with readout noise and scattering.

```python
# -----------------------------------
# Plotting function
# -----------------------------------
def plot_rmf_set1(E_channel, rmf_dist, mormf_dist, Ein, pdf):
    gs = gridspec.GridSpec(2, 1, height_ratios=[2, 0.5], hspace=0)
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1], sharex=ax0)

    # --- Main RMF plot (log-log scale) ---
    ax0.plot(E_channel, rmf_dist, 'o-', color='k', ms=1.5, lw=1, alpha=0.8, label = f"SXT rmf @ {Ein:.4f} keV")
    ax0.plot(E_channel, mormf_dist, color='tomato', lw=2, label=f'model rmf Gen1; {Bt:.4f}, {f0:.4f}')
    ax0.set_xscale('log')
    ax0.set_yscale('log')
    ax0.set_xlim(0.05, 2.2)
    ax0.set_ylim(1e-6, 1e-1)
    ax0.set_ylabel('N(x) distribution')
    ax0.legend()
    ax0.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    ax0.set_title(f'SXT RMF model for X-ray Photon Energy {float(Ein):.4f} keV')

    # --- Residual plot (Data / Model, log-log scale) ---
    ax1.plot(E_channel, rmf_dist / mormf_dist, '-.', color='tomato', lw=0.8, label='Ratio: Data / Model')
    ax1.axhline(1, color='black', linestyle='--', linewidth=0.6)
    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.set_xlim(0.05, 2.2)
    ax1.set_ylim(0, 15)
    ax1.set_xlabel('Channel Energy (E) keV')
    ax1.set_ylabel('Data/Model')
    ax1.legend(prop={'size': 5})
    ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

    plt.tight_layout()
    pdf.savefig()
    plt.close()


def plot_rmf_set2(E_channel, rmf_dist, mormf_dist, Energy, mfit_y, Ein, pdf):
    gs = gridspec.GridSpec(2, 1, height_ratios=[2, 0.5], hspace=0)
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1], sharex=ax0)

    ax0.plot(E_channel, rmf_dist, 'o-', color='k', ms=1.5, lw=1, alpha=0.8, label=f'SXT rmf @ {float(Ein):.4f} keV')
    ax0.plot(E_channel, mormf_dist, color='tomato', lw=2, label='model rmf Gen1')
    ax0.plot(Energy, mfit_y, color='darkblue', lw=0.4, label='check')

    ax0.set_xlim(0, 11)
    ax0.set_ylabel('N(x) distribution')
    ax0.legend()
    ax0.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    ax0.set_title(f'SXT RMF model for X-ray Photon Energy {Ein:.4f} keV')

    ax1.plot(E_channel, rmf_dist - mormf_dist, '-.', color='tomato', lw=0.8, label='Residual: original - model')
    ax1.axhline(0, color='black', linestyle='--', linewidth=0.6)
    ax1.set_xlim(0, 11)
    ax1.set_xlabel('Channel Energy (E) keV')
    ax1.set_ylabel('Residual')
    ax1.legend(prop={'size': 5})
    ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

    plt.tight_layout()
    pdf.savefig()
    plt.close()
```

```python
# -----------------------------------
# Main loop
# -----------------------------------

PlotDir = './Updated_RMF/Modified_RMF_plots/'
pdf_filename = os.path.join(PlotDir, 'Modified_RMF_test.pdf')

RmfDir ='./input/RMF_G0to4/'
paramsDir = './input/params/'

with PdfPages(pdf_filename) as pdf:
    for i in range(285):
        idx = 83 + i

        # Load original RMF to calibrate
        
        RMFg0to4= np.loadtxt(os.path.join(RmfDir, f'rmf_{idx}.dat'))
        E_channel,  RMFDist   = RMFg0to4[:, 0]* 1e-2 , RMFg0to4[:, 1]
        data2 = np.loadtxt(f'{rootdir6}/rmf_{idx}.dat')
#        E_channel = data2[:, 0] * 1e-2  # convert to keV
#        rmf_dist = data2[:, 1]

        # Load attenuation coefficient and Ein
        AttCoeff = np.loadtxt(os.path.join(paramsDir, 'attcoeff.dat'))
        Ein, mu = AttCoeff[:, 0][idx], AttCoeff[:, 1][idx]

        print('Modeling RMF at X-ray Photon Energy:', np.round(Ein,2))

        # Load Parameters values
        l = 1.74
        param_tau  =  np.loadtxt(os.path.join(paramsDir, 'tau_val.dat'))
        tau = param_tau[:, 1][i]
        param_sigma = np.loadtxt(os.path.join(paramsDir, 'sigma_val.dat'))
        sig = param_sigma[:, 1][i]
        param_norm = np.loadtxt(os.path.join(paramsDir, 'norm_val.dat'))
        norm = param_norm[:, 1][i] 
        
        data4 = np.loadtxt(f'{rootdir3}/params_set1.dat')
        f0   = data4[:, 9][i]

        Bt   = data4[:, 3][i]
        
        # Compute model rmf distribution
        mormfdist1 = mormf_set1(E_channel, l, Bt, sig, tau, norm, f0, mu, Ein)

        np.savetxt(rootdir4 + f'mormf_{idx}.dat', np.column_stack([E_channel, mormfdist1, RMFDist - mormfdist1]), fmt='%.20g')

        # Plot and save
        plot_rmf_set1(E_channel, RMFDist, mormfdist1, Ein, pdf)
        matrix_data[idx][5] = mormfdist1

print(f"Plots saved to {pdf_filename}")


# âœï¸ Modify MATRIX header
matrix_header = file['MATRIX'].header
matrix_header['CDES0001'] = 'SXT RMF regenerated from Model'
matrix_header['DATE'] = '2025-05-13'  

# Save updated FITS file
file.writeto(OutputRMF, overwrite=True)
file.close()

print(f"Modified RMF saved as: {OutputRMF}")
```
